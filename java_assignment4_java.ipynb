```python
%%writefile Main.java
// =========================
// âœ… ALL JAVA PROGRAMS + ANSWERS COMBINED
// =========================

// =========================
// GENERICS SECTION
// =========================

// 1. Purpose of Generics
// -> Provides type safety & code reusability (Compile-time type checking, avoids ClassCastException)

// 2. User-defined generic class
class MyGeneric<T> {
    private T value;
    public MyGeneric(T value){ this.value = value; }
    public T getValue(){ return value; }
}

// 3. Bounded type parameter (only Number subclasses allowed)
class NumericBox<T extends Number> {
    private T num;
    public NumericBox(T num){ this.num = num; }
    public T getNum(){ return num; }
}

// 4. ? extends T vs ? super T
// ? extends T  => read-only (producer)
// ? super T    => write allowed (consumer)

// 5. Raw type vs parameterized type
// Raw type: MyGeneric g = new MyGeneric("hello");  // unsafe
// Parameterized: MyGeneric<String> g = new MyGeneric<>("hello"); // safe

// 6. Generic Pair class
class Pair<K, V> {
    private K key;
    private V value;
    public Pair(K key, V value){ this.key = key; this.value = value; }
    public K getKey(){ return key; }
    public V getValue(){ return value; }
    public void setKey(K key){ this.key = key; }
    public void setValue(V value){ this.value = value; }
}

// 7. Generic Box class
class Box<T> {
    private T item;
    public void addItem(T item){ this.item = item; }
    public T getItem(){ return item; }
}

// 8. Generic swap method
class Utils {
    public static <T> void swapElements(T[] arr, int i, int j){
        T temp = arr[i];
        arr[i] = arr[j];
        arr[j] = temp;
    }
}

// 9. Generic Stack
class Stack<T> {
    private java.util.ArrayList<T> list = new java.util.ArrayList<>();
    public void push(T item){ list.add(item); }
    public T pop(){ return list.remove(list.size()-1); }
    public T peek(){ return list.get(list.size()-1); }
    public boolean isEmpty(){ return list.isEmpty(); }
}

// 10. MinMaxFinder
class MinMaxFinder<T extends Comparable<T>> {
    private java.util.List<T> list;
    public MinMaxFinder(java.util.List<T> list){ this.list = list; }
    public T findMin(){ return java.util.Collections.min(list); }
    public T findMax(){ return java.util.Collections.max(list); }
}

// =========================
// COLLECTION FRAMEWORK
// =========================
// (Other answers explained via comments + demo code in main)

// =========================
// MAIN METHOD (RUN DEMOS)
// =========================
public class Main {
    public static void main(String[] args) {

        // ===== Demo for Box<T> =====
        Box<String> box1 = new Box<>();
        box1.addItem("Hello");
        System.out.println("Box1 String: " + box1.getItem());

        Box<Integer> box2 = new Box<>();
        box2.addItem(100);
        System.out.println("Box2 Integer: " + box2.getItem());

        // ===== Demo for swapElements =====
        Integer[] nums = {1, 2, 3, 4};
        Utils.swapElements(nums, 0, 3);
        System.out.println("Swapped Array: " + java.util.Arrays.toString(nums));

        // ===== Demo for Stack<T> =====
        Stack<String> stack = new Stack<>();
        stack.push("A");
        stack.push("B");
        System.out.println("Stack Peek: " + stack.peek());
        System.out.println("Stack Pop: " + stack.pop());

        // ===== Demo for MinMaxFinder =====
        java.util.List<Integer> intList = java.util.Arrays.asList(10, 20, 5, 40);
        MinMaxFinder<Integer> mm = new MinMaxFinder<>(intList);
        System.out.println("Min: " + mm.findMin());
        System.out.println("Max: " + mm.findMax());
    }
}
```
