```python
%%writefile Module4_Assignment.java
// ===============================================
// MODULE 4 ASSIGNMENT â€” JAVA GENERICS & COLLECTIONS
// Sparsh Goyal
// ===============================================

import java.util.*;

// MAIN class to execute demos
public class Module4_Assignment {
    public static void main(String[] args) {

        System.out.println("\n===== 1. GENERICS =====");

        // --- Box<T> demo ---
        Box<String> strBox = new Box<>();
        strBox.addItem("Hello Generic World!");
        System.out.println("Box<String>: " + strBox.getItem());

        Box<Integer> intBox = new Box<>();
        intBox.addItem(42);
        System.out.println("Box<Integer>: " + intBox.getItem());

        // --- Pair<K,V> demo ---
        Pair<String, Integer> student = new Pair<>("Sparsh", 90);
        System.out.println("Pair: " + student.getKey() + " -> " + student.getValue());

        // --- SwapElements demo ---
        Integer[] arr = {1, 2, 3};
        Utils.swapElements(arr, 0, 2);
        System.out.println("Swapped Array: " + Arrays.toString(arr));

        // --- Stack<T> demo ---
        MyStack<String> s = new MyStack<>();
        s.push("A"); s.push("B");
        System.out.println("Stack Peek: " + s.peek());
        System.out.println("Stack Pop: " + s.pop());

        // --- MinMaxFinder demo ---
        MinMaxFinder<Integer> mm = new MinMaxFinder<>(List.of(3, 1, 5, 2));
        System.out.println("Min: " + mm.findMin() + " | Max: " + mm.findMax());


        System.out.println("\n===== 2. COLLECTION FRAMEWORK =====");

        // --- Iterate list ---
        List<Integer> nums = Arrays.asList(1, 2, 3);
        System.out.print("For loop: ");
        for (int i = 0; i < nums.size(); i++) System.out.print(nums.get(i) + " ");
        System.out.print("\nEnhanced for: ");
        for (int n : nums) System.out.print(n + " ");
        System.out.print("\nIterator: ");
        Iterator<Integer> it = nums.iterator();
        while (it.hasNext()) System.out.print(it.next() + " ");
        System.out.println();

        // --- Generic printCollection ---
        Utils.printCollection(Set.of("X", "Y", "Z"));


        System.out.println("\n===== 3. LIST INTERFACE =====");

        // --- ArrayList basic ops ---
        List<String> list = new ArrayList<>(List.of("alpha", "beta", "gamma"));
        System.out.println("Initial: " + list);
        list.remove("beta");
        System.out.println("After remove: " + list);
        list.add(0, "new");
        list.set(1, "replaced");
        System.out.println("After updates: " + list);

        // --- Sort demo ---
        Collections.sort(list);
        System.out.println("Sorted: " + list);
        Collections.sort(list, Collections.reverseOrder());
        System.out.println("Reverse sorted: " + list);


        System.out.println("\n===== 4. SET INTERFACE =====");

        Set<String> hs = new HashSet<>(List.of("apple", "banana", "apple"));
        System.out.println("HashSet (duplicates ignored): " + hs);

        Set<String> lhs = new LinkedHashSet<>(List.of("one", "two", "three"));
        System.out.println("LinkedHashSet (order preserved): " + lhs);

        TreeSet<Integer> ts = new TreeSet<>(List.of(5, 2, 8, 1));
        System.out.println("TreeSet sorted: " + ts);
        System.out.println("Smallest: " + ts.first() + " | Largest: " + ts.last());


        System.out.println("\n===== 5. MAP INTERFACE =====");

        Map<Integer, String> emp = new HashMap<>();
        emp.put(101, "Alice");
        emp.put(102, "Bob");
        System.out.println("HashMap: " + emp);

        System.out.print("Iterate via KeySet: ");
        for (int k : emp.keySet()) System.out.print(k + "=" + emp.get(k) + " ");
        System.out.print("\nIterate via EntrySet: ");
        for (Map.Entry<Integer, String> e : emp.entrySet())
            System.out.print(e.getKey() + ":" + e.getValue() + " ");
        System.out.println();

        TreeMap<String, Integer> tm = new TreeMap<>();
        tm.put("z", 1); tm.put("a", 2); tm.put("m", 3);
        System.out.println("TreeMap (sorted keys): " + tm);

        Map<String, Integer> lhm = new LinkedHashMap<>();
        lhm.put("b", 2); lhm.put("a", 1); lhm.put("c", 3);
        System.out.println("LinkedHashMap (insertion order): " + lhm);


        System.out.println("\n===== 6. QUEUE & STACK =====");

        // Queue demo
        Queue<String> q = new LinkedList<>();
        q.add("User1"); q.add("User2"); q.add("User3");
        while (!q.isEmpty()) System.out.println("Serving: " + q.poll());

        // PriorityQueue demo
        PriorityQueue<Task> pq = new PriorityQueue<>(Comparator.comparingInt(Task::getPriority).reversed());
        pq.add(new Task("Low", 1)); pq.add(new Task("High", 10)); pq.add(new Task("Mid", 5));
        System.out.println("Top priority task: " + pq.poll());

        // Deque demo
        ArrayDeque<String> dq = new ArrayDeque<>();
        dq.addFirst("front");
        dq.addLast("back");
        System.out.println("Deque: " + dq);
        dq.removeFirst(); dq.removeLast();
        System.out.println("Deque after removals: " + dq);

        // Palindrome check
        System.out.println("Is 'level' palindrome? " + Utils.isPalindrome("level"));


        System.out.println("\n===== 7. CONCURRENCY DEMO =====");

        // Vector thread-safe demo
        Vector<Integer> v = new Vector<>();
        Runnable r = () -> { for (int i = 0; i < 1000; i++) v.add(i); };
        Thread t1 = new Thread(r);
        Thread t2 = new Thread(r);
        t1.start(); t2.start();
        try { t1.join(); t2.join(); } catch (InterruptedException e) {}
        System.out.println("Vector size after threads: " + v.size());

        // CopyOnWriteArrayList demo
        java.util.concurrent.CopyOnWriteArrayList<String> cow = new java.util.concurrent.CopyOnWriteArrayList<>();
        cow.add("A"); cow.add("B");
        for (String s1 : cow) {
            if (s1.equals("A")) cow.add("C");
        }
        System.out.println("CopyOnWriteArrayList safe modify: " + cow);


        System.out.println("\n===== 8. COLLECTIONS UTILITY =====");

        List<String> fruits = new ArrayList<>(List.of("apple", "banana", "cherry"));
        Collections.shuffle(fruits);
        System.out.println("Shuffled: " + fruits);
        Collections.sort(fruits);
        System.out.println("Sorted: " + fruits);
        System.out.println("Binary search 'banana': " + Collections.binarySearch(fruits, "banana"));
        System.out.println("Frequency of 'apple': " + Collections.frequency(fruits, "apple"));

        List<String> unmod = Collections.unmodifiableList(fruits);
        System.out.println("Unmodifiable list: " + unmod);
        try { unmod.add("new"); } catch (UnsupportedOperationException e) {
            System.out.println("Modification not allowed!");
        }

        System.out.println("\n===== END OF MODULE 4 DEMO =====");
    }
}


// ===================== SUPPORT CLASSES =====================

// Generic Box<T>
class Box<T> {
    private T item;
    public void addItem(T item) { this.item = item; }
    public T getItem() { return item; }
}

// Generic Pair<K,V>
class Pair<K,V> {
    private K key;
    private V value;
    public Pair(K key, V value) { this.key = key; this.value = value; }
    public K getKey() { return key; }
    public V getValue() { return value; }
    public void setKey(K k) { key = k; }
    public void setValue(V v) { value = v; }
}

// Generic Stack<T>
class MyStack<T> {
    private ArrayList<T> data = new ArrayList<>();
    public void push(T item) { data.add(item); }
    public T pop() { return data.remove(data.size() - 1); }
    public T peek() { return data.get(data.size() - 1); }
    public boolean isEmpty() { return data.isEmpty(); }
}

// Generic MinMaxFinder<T extends Comparable<T>>
class MinMaxFinder<T extends Comparable<T>> {
    private List<T> list;
    public MinMaxFinder(List<T> list) { this.list = list; }
    public T findMin() {
        if (list == null || list.isEmpty()) return null;
        T min = list.get(0);
        for (T e : list) if (e.compareTo(min) < 0) min = e;
        return min;
    }
    public T findMax() {
        if (list == null || list.isEmpty()) return null;
        T max = list.get(0);
        for (T e : list) if (e.compareTo(max) > 0) max = e;
        return max;
    }
}

// Utility helper methods
class Utils {
    public static <T> void swapElements(T[] arr, int i, int j) {
        T temp = arr[i]; arr[i] = arr[j]; arr[j] = temp;
    }
    public static <T> void printCollection(Collection<T> c) {
        System.out.print("Collection: ");
        for (T e : c) System.out.print(e + " ");
        System.out.println();
    }
    public static boolean isPalindrome(String s) {
        Deque<Character> d = new ArrayDeque<>();
        for (char c : s.toCharArray()) d.addLast(c);
        while (d.size() > 1) {
            if (!d.removeFirst().equals(d.removeLast())) return false;
        }
        return true;
    }
}

// Task class for PriorityQueue demo
class Task {
    private String name;
    private int priority;
    public Task(String name, int priority) { this.name = name; this.priority = priority; }
    public int getPriority() { return priority; }
    public String toString() { return name + " (Priority " + priority + ")"; }
}
```
